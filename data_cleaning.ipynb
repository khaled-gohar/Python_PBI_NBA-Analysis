```python
import pandas as pd
```

```python
mvp = pd.read_csv('mvps.csv')
```

```python
mvp
```

```python
mvp = mvp[['Player','Pts Won','Pts Max','Share','Year']]
```

```python
players = pd.read_csv('players.csv')
```

```python
players.head(20)
```

```python
del players['Unnamed: 0']
del players['Rk']
```

```python
players['Player'] = players['Player'].str.replace('*','',regex=False)
```

```python
players.groupby(['Player','Year']).get_group(('Greg Anderson',1991))
```

```python
def sigle_row(df):
    if df.shape[0] == 1:
        return df
    else:
        row = df[df['Tm'] == 'TOT']
        row['Tm'] = df.iloc[-1,:]['Tm']
        return row
    
players=players.groupby(['Player','Year']).apply(sigle_row)
```

```python
players
```

```python
players.index = players.index.droplevel()
```

```python
players.index = players.index.droplevel()
```

```python
players
```

```python
combined = players.merge(mvp,how='outer',on=['Player','Year'])
```

```python
combined[['Pts Won','Pts Max','Share']] = combined[['Pts Won','Pts Max','Share']].fillna(0)
```

```python
combined.head(50)
```

```python
teams = pd.read_csv('teams.csv',encoding='cp1252')
```

```python
teams.head(30)
```

```python
teams = teams[~teams['W'].str.contains('Division')]
```

```python
teams['Team']= teams['Team'].str.replace('*','',regex=False)
```

```python
teams['Team'] = teams['Team'].str.replace('(','',regex=False)
teams['Team'] = teams['Team'].str.replace(')','',regex=False)
```

```python
teams['Team'] = teams['Team'].str.replace('\d+','',regex=True)
```

```python
teams['Team'] =teams['Team'].str.replace('\xa0','',regex=False)
```

```python
nicknames= {}

with open('nicknames.csv') as f:
    lines = f.readlines()
    for line in lines:
        abbrev,name = line.replace('\n','').split(',')
        nicknames[abbrev] = name

```

```python
nicknames
```

```python
combined['Team']  = combined['Tm'].map(nicknames)
```
